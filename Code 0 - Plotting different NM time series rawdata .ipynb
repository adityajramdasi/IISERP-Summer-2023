## About

This is the first code I wrote for this project. Here, I simply extract the raw count/sec NM data (24 hours before shock arrival start till 24 hours after MC end) with 1hour resolution from the downloaded csv file, manually sorting it out according to stations. I do this for 20 ICME events, the first of every year from 2015-2009, 2004, 2001, as identified in Debesh Bhattacharjee's paper (appendix). I also plot the time stamps for shock arrival, MC start and MC end from the WIND catalogue onto the NM time series plots.
